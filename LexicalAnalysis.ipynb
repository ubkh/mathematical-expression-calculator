```python
class Token:
    def __init__(self, type, value):
        self.type = type
        self.value = value
        self.associativity = None
        self.precedence = None
```

```python

class Stack:
    '''Data structure for a Stack (LIFO)'''

    def __init__(self, another=None):
        self.list = []
        if another:
            self.list = another.list

    def push(self, val):
        self.list.append(val)

    def pop(self):
        return self.list.pop() if self.list else None

    def peek(self):
        return self.list[-1] if self.list else None

    def isEmpty(self):
        return False if self.list else True

    def size(self):
        return len(self.list)

    def display(self):
        #print(self.list)
        
        #prints list elements on each line rather than whole list at once
        for element in range(0,len(self.list)):
            print(self.list[element])
            print(self.list[element].type)
            print(self.list[element].value)
            
        return self.list
```

```python
def isDigit(char):
    pattern = re.compile("\d+(\.\d*)?")
    if(pattern.match(char)):
        return True

def isLetter(char):
    pattern = re.compile("[a-zA-Z]")
    if(pattern.match(char)):
        return True

def isOperator(char):
    pattern = re.compile("[+\-\*\/^]")
    if(pattern.match(char)):
        return True

def isLeftParenthesis(char):
    return char == "("

def isRightParenthesis(char):
    return char == ")"

def isMinus(char):
    pattern = re.compile("[-]")
    if(pattern.match(char)):
        return True
    else:
        return False
```

```python
def Tokenise(str):
    result = Stack()
    
    unaryFound = False
    
    numBuffer = []
    letterBuffer = []
    operatorBuffer = []
    
    #remove all superfluous whitespace
    str.replace(" ", "")
    str = ''.join(str.split())
    
    #convert string into array of chars
    str = list(str)

    for index, char in enumerate(str):
        if(isDigit(char)):
            if(unaryFound == True):
                numBuffer.append(operatorBuffer[-1] + char)
                unaryFound = False
            else:
                numBuffer.append(char)
            
            operatorBuffer.clear()

        elif(char=="."):
            numBuffer.append(char)
            
        elif(isLetter(char)):
            if numBuffer:
                numBuffContents = ''.join(numBuffer)
                result.push(Token("Literal", numBuffContents))
                #empty out the numBuffer
                numBuffer.clear()
                
                
            if(unaryFound == True):
                result.push(Token("Literal", operatorBuffer[-1] + "1"))
                operatorBuffer.clear()
                unaryFound = False
                
            letterBuffer.append(char)

        elif(isOperator(char)):
            operatorBuffer.append(char)
            
            #if numBuffer empty
            if(len(numBuffer) == 0) and (len(letterBuffer) == 0) and (len(operatorBuffer) == 1):
                if(isMinus(operatorBuffer[-1]) and result.peek() and result.peek().value == ")"):
                    result.push(Token("Operator", char))
                elif(isMinus(operatorBuffer[-1]) == False and result.peek() and result.peek().value == ")"):
                    result.push(Token("Operator", char))
                else:
                    unaryFound = True
                    
            elif(len(operatorBuffer) > 1):
                if(isMinus(operatorBuffer[-1])):
                    unaryFound = True
   
            #if numBuffer is not empty
            elif numBuffer:
                numBuffContents = ''.join(numBuffer)
                result.push(Token("Literal", numBuffContents))
                numBuffer.clear()
                result.push(Token("Operator", char))
            else:
                for elementNum, element in enumerate(letterBuffer):
                    if result.isEmpty() == False:
                        result.push(Token("Operator", "*"))
                    result.push(Token("Variable", element))
                letterBuffer.clear()
                result.push(Token("Operator", char))
                
        elif(isLeftParenthesis(char)):
            if(index!=0) and (result.size()>0) and (len(operatorBuffer)==0):
                result.push(Token("Operator", "*"))
            
            operatorBuffer.clear()
            
            letterBuffContents = ''.join(letterBuffer)
            if len(letterBuffContents) > 0:
                result.push(Token("Function", letterBuffContents))
                
            letterBuffer.clear()
            result.push(Token("Left Parenthesis", char))
        elif(isRightParenthesis(char)):
            operatorBuffer.clear()
                 
            if numBuffer:
                numBuffContents = ''.join(numBuffer)
                result.push(Token("Literal", numBuffContents)) 
                numBuffer.clear()
                
            for element in letterBuffer:
                if(len(letterBuffer)>0):
                    if(result.peek() and result.peek().type != "Operator"):
                        result.push(Token("Operator", "*"))
                result.push(Token("Variable", element))
                
            letterBuffer.clear()
            result.push(Token("Right Parenthesis", char))
    
    if numBuffer:
        numBuffContents = ''.join(numBuffer)
        result.push(Token("Literal", numBuffContents))
        numBuffer.clear()
        
    if letterBuffer:
        for element in letterBuffer:
            result.push(Token("Operator", "*"))
            result.push(Token("Variable", element))
        letterBuffer.clear()
    
    #currently only displays list of token objects
    return result.display()
```
